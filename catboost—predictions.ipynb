```python
import random
def fit_over_sample(train_data):
    """Resample the dataset.
    """
    y = train_data.hand
    label = np.unique(y)
    stats_c_ = {}
    maj_n = 0
    for i in label:
        nk = sum(y==i)
        stats_c_[i] = nk
        if nk > maj_n:
            maj_n = nk 
            maj_c_ = i
 
 
    # Keep the samples from the majority class
    X_resampled = train_data[y == maj_c_]
    print('maj_c: {}'.format(len(X_resampled)))
    # Loop over the other classes over picking at random
    for key in stats_c_.keys():
 
        # If this is the majority class, skip it
        if key == maj_c_:
            continue
 
        # Define the number of sample to create
        num_samples = int(stats_c_[maj_c_] -stats_c_[key])
        index_samples = train_data[y==key].index.get_values()
        print('num_sample:{}'.format(num_samples))
        new_resampled = train_data[y==key]
        print('new_resampled_pre:{}'.format(len(new_resampled)))
        for i in range(num_samples):
            new_resampled = pd.concat([new_resampled, train_data.iloc[random.sample(index_samples, 1)]],axis=0)
        print('after:{}'.format(len(new_resampled)))
        X_resampled = pd.concat([X_resampled, new_resampled], axis=0)
    return X_resampled
 
 
 
# Apply the random over-sampling
train= fit_over_sample(training)
```

```python
def bincount2D_vectorized(a):    
    N = a.max()+1
    a_offs = a + np.arange(a.shape[0])[:,None]*N
    return np.bincount(a_offs.ravel(), minlength=a.shape[0]*N).reshape(-1,N)

S_training = training.iloc[:,[0,2,4,6,8]].astype(int)
S_training = pd.DataFrame(bincount2D_vectorized(S_training.values),columns=['suitCount0','suitCount1','suitCount2','suitCount3','suitCount4'])
training = pd.merge(training, S_training, how='left', left_index=True, right_index=True).drop(['suitCount0'],axis=1)


R_training = training.iloc[:,np.arange(1,10,2)].astype(int)
cols = ['rank{}'.format(x) for x in range(0,14,1)]
R_training = pd.DataFrame(bincount2D_vectorized(R_training.values),columns=cols)
training = pd.merge(training, R_training, how='left', left_index=True, right_index=True).drop(['rank0'],axis=1)

training.head()
```

```python
#----------------------


X = training.drop(['hand'],axis=1)
y = training.hand

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss
import catboost as cb

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
params = {
            'l2_leaf_reg':0.8,
            'learning_rate':0.09,
            'depth':11,
            'iterations':250
         }
cat = cb.CatBoostClassifier(loss_function='MultiClassOneVsAll', random_seed=1234)
cat.set_params(**params)
cat_model = cat.fit(X_train, y_train, verbose=False) 
preds_class = cat_model.predict(X_test, prediction_type='Class')
preds_proba = cat_model.predict(X_test, prediction_type='Probability')

print("CV accuracy: {}".format(accuracy_score(y_test,preds_class)))
print("CV logloss: {}".format(log_loss(y_test,preds_proba)))
```
